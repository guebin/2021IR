```r
#1
y = c()
t = c()
for (i in 1:1000){
  t[i] = (2*pi*i)/1000
  y[i] = 1.5+(5*sin(t[i]))+(3*cos(4*t[i]))
}
```

```r
#1-1
epsilon = rnorm(1000,mean=0,sd=1)
epsilon
```

```r
y = y+epsilon
```

```r
#1-2
x_1 = c()
x_2 = c()
for (i in 1:1000){
  x_1[i] = sin(t[i])
  x_2[i] = cos(4*t[i])
}
```

```r
#1-3. 분명 모양은 같은데 왜 다른 그래프 같을까요..ㅠ rnorm때문이겠죠..?
for (i in 1:1000){
  y[i] = 1.5+(5*x_1[i])+(3*x_2[i])
}
y = y+epsilon
plot(t,y,col='gray60')
```

```r
#1-4
X = cbind(1,x_1,x_2)
X
```

```r
#1-5
beta = rbind(1.5,5,3)
rslt = X %*% beta

for (i in 1:1000){
  y[i] = 1.5+(5*x_1[i])+(3*x_2[i])
}
y = y+epsilon
plot(t,y,col='gray60')
lines(t,rslt,col='red',lwd=5)
```

```r
#1-6. 거의 같습니다!
betahat = solve(t(X) %*% X) %*% t(X) %*% y
print(beta)
betahat
```

```r
#2-1
x=seq(from=-1.96,to=1.96,by=0.01)
y=(1/sqrt(2*pi))*exp(-1/2*x^2)
xx = runif(10000, -2, 2)
yy = runif(10000, 0.05, 0.4)
plot(xx,yy)
lines(x,y,col='red',lwd=5)
```

```r
test = function(xx,yy) {
  yy < (1/sqrt(2*pi))*exp(-1/2*xx^2)
}
tst = c()
for (i in 1:10000) tst[i] = test(xx[i],yy[i])
plot(xx,yy,col='gray')
lines(x,y,col='red',lwd=5)
points(xx[tst],yy[tst],col='red')
```

3 : 10명의 참가자, 유리쌍 20개(총 40장), 9번은 유리장인으로 95%, 1~8,10은 일반인으로 50%
- 내림차순(10부터 1), 오름차순(1부터 10) 2가지 방법이 있음.
- 8은 어떻게 해야 생존확률이 더 높나

```r
#3. Type A
ARR = c('N1','N2','N3','N4','N5','N6','N7','N8','A','N9')
SURV = 10
PLAYER = ARR[SURV]
STAGE = 0
tossRSLT = NA

reset = function(){
SURV <<- 10
STAGE <<- 0
PLAYER <<- ARR[SURV]
tossRSLT <<- NA
}
toss = function(p) rbinom(n=1, size=1, prob=p) %>% as.logical
go = function(){
  for (i in 1:20){
    PROB = 0.5+(PLAYER=='A')*0.45
    tossRSLT <<- toss(PROB)
    if (tossRSLT == FALSE) SURV <<- SURV-1
    STAGE <<- STAGE + 1
    PLAYER <<- ARR[SURV]
    if (SURV == 0) break
  }
}
record = function(){
  list(SURV=SURV,STAGE=STAGE, PLAYER=PLAYER)
}
simulate_once = function(){
  reset()
  go()
  return(record()$SURV)
}

record = function(){
  list(SURV=SURV, STAGE=STAGE)
  }

surv8 = 0
for (i in 1:10000){
  reset()
  go()
  record()
  if (record()$SURV > 7) surv8 <<- surv8 + 1
}

print(surv8/10000)
```

```r
#3.Type B
ARR = c('N9','A','N8','N7','N6','N5','N4','N3','N2','N1')
SURV = 10
PLAYER = ARR[SURV]
STAGE = 0
tossRSLT = NA

reset = function(){
SURV <<- 10
STAGE <<- 0
PLAYER <<- ARR[SURV]
tossRSLT <<- NA
}
toss = function(p) rbinom(n=1, size=1, prob=p) %>% as.logical
go = function(){
  for (i in 1:20){
    PROB = 0.5+(PLAYER=='A')*0.45
    tossRSLT <<- toss(PROB)
    if (tossRSLT == FALSE) SURV <<- SURV-1
    STAGE <<- STAGE + 1
    PLAYER <<- ARR[SURV]
    if (SURV == 0) break
  }
}
record = function(){
  list(SURV=SURV,STAGE=STAGE, PLAYER=PLAYER)
}
simulate_once = function(){
  reset()
  go()
  return(record()$SURV)
}

record = function(){
  list(SURV=SURV, STAGE=STAGE)
  }

surv8 = 0
for (i in 1:10000){
  reset()
  go()
  record()
  if (record()$SURV <= 3) surv8 <<- surv8 + 1
}

print(surv8/10000)
```
