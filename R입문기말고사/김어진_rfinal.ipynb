(4) p=0.2인 베르누이분포에서 1000개 난수를 만들면 1 몇개 포함되는지 시뮬레이션

```r
simulate_once=function(){
    rbinom(1000, size=1, prob=0.2) %>% sum()
}

simrslt=c()

for(i in 1:100000) simrslt[i]=simulate_once()
mean(simrslt)
```
